```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cats_bgr = cv2.imread("/content/drive/MyDrive/goat.jpeg")
cats_rgb = cv2.cvtColor(cats_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cats_rgb)
plt.axis('off')
plt.show()
```

HELPER FUNCTION


```python
def apply_filters(kernel, image):

  filtered_img = cv2.filter2D(src=image, kernel=kernel, ddepth=-1)
  plt.figure(figsize=(8,8))
  plt.imshow(filtered_img, cmap='gray')
  plt.axis('off')
  plt.show()
```

## Learn and apply filters!

```python
#Blur filter

blur_kernel = np.ones((9, 9)) / 81
apply_filters(kernel=blur_kernel, image=cats_rgb.copy())
```

```python
#sepia filter

sepia_kernel = np.asarray([[0.393, 0.769, 0.189],
                           [0.349, 0.686, 0.168 ],
                            [0.272, 0.534, 0.131]])

apply_filters(kernel=sepia_kernel, image=cats_rgb.copy())
```

```python
#emboss filter

emboss_kernel = np.array([[-6, -1, 0],
                         [-1, 1, 1],
                         [0, 1, 6]])
apply_filters(kernel=emboss_kernel, image=cats_rgb.copy())
```

```python
#Edge detection

edge_horizontal_kernel = np.array([[-1, 0, 1],
                                   [-2, 0, 2],
                                   [-1, 0, 1]])

apply_filters(kernel=edge_horizontal_kernel, image=cats_rgb.copy())
```

```python
edge_vertical_kernel = np.array([[-1, -2, -1],
                                   [0, 0, 0],
                                   [1, 2, 1]])

apply_filters(kernel=edge_vertical_kernel, image=cats_rgb.copy())
```
